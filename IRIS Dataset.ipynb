```python
import seaborn as sns
import matplotlib.pyplot as plt

sub = iris[iris['Species'] == 'Iris-setosa']

sns.kdeplot(
    x=sub['SepalLengthCm'], 
    y=sub['SepalWidthCm'], 
    fill=True,        # equivalent to `shade=True`
    cmap="plasma",    # works now with `fill=True`
    thresh=0,         # controls minimum density threshold
    levels=100        # optional: number of contour levels
)

plt.title('Iris-setosa')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sub = iris[iris['Species'] == 'Iris-setosa']

sns.kdeplot(
    x=sub['SepalLengthCm'],
    y=sub['SepalWidthCm'],
    fill=True,
    cmap='plasma',
    thresh=0,
    levels=100
)

plt.title('Iris-setosa')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
